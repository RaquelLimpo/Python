## LoginPY
***
Escribe un programa que solicite al usuario ingresar un nombre de usuario y una contraseña. El usuario tendrá inicialmente 3 intentos para acceder al sistema.  <br><br>
• **CASO 1**: Si el nombre de usuario introducido es "admin" y la contraseña es "admin123", el programa mostrará el mensaje: *“Has accedido al 
sistema como Admin”*.  <br>
• **CASO 2**: Si el nombre de usuario es “dporti” y la contraseña es “1234”, el programa mostrará el mensaje: *“Bienvenido al sistema, David Porti” *.  <br>
• **CASO 3**: Si el nombre de usuario es “guest” y la contraseña es “guestpass”, el programa mostrará el mensaje:* “Bienvenido al sistema en modo invitado”*.  <br> 
• **CASO 4**: Si el nombre de usuario es “newuser” y la contraseña es “newpass”, el programa mostrará el mensaje: *“Cuenta creada correctamente para Nuevo Usuario”. *  <br>
• **Cualquier otro caso**, se mostrará un mensaje de error indicando: *“No es posible acceder con las credenciales proporcionadas”* junto con la cantidad de intentos restantes. El programa volverá a solicitar el nombre de usuario y contraseña hasta un máximo de 3 intentos.  <br>
El programa limpiará la pantalla antes de solicitar el nombre de usuario y contraseña en cada intento.  <br>
Si el usuario excede los 3 intentos, el programa mostrará el mensaje: “Programa bloqueado” y finalizará. <br>

```python
import os

def limpiar_pantalla():
    os.system('cls' if os.name == 'nt' else 'clear')

def login():
    intentos = 3

    while intentos > 0:
        limpiar_pantalla()
        print(f"Tienes {intentos} intento(s) restante(s).\n")

        usuario = input("Ingrese el nombre de usuario: ").strip()
        contrasena = input("Ingrese la contraseña: ").strip()

        if usuario == "admin" and contrasena == "admin123":
            print("\nHas accedido al sistema como Admin.")
            return
        elif usuario == "dporti" and contrasena == "1234":
            print("\nBienvenido al sistema, David Porti.")
            return
        elif usuario == "guest" and contrasena == "guestpass":
            print("\nBienvenido al sistema en modo invitado.")
            return
        elif usuario == "newuser" and contrasena == "newpass":
            print("\nCuenta creada correctamente para Nuevo Usuario.")
            return
        else:
            intentos -= 1
            print(f"\nNo es posible acceder con las credenciales proporcionadas. Intentos restantes: {intentos}")

    limpiar_pantalla()
    print("\nPrograma bloqueado.")

if __name__ == "__main__":
    login()

```

## AreaCirculoPY 
***
 Escribe un programa que solicite al usuario el radio de un círculo.<br>
 A continuación, el programa deberá calcular y mostrar el área del círculo. <br>
 El valor introducido debe considerarse con decimales y el resultado debe expresarse con dos decimales. <br>
Para realizar el cálculo del área, debes crear un método llamado calculoArea dentro del fichero llamado AreaCirculoPY, el cual recibirá el radio como parámetro.  <br>
Este método debe ser colocado en un archivo llamado AreaCirculo.py. No hace falta en un fichero externo. <br>
Se debe verificar que el radio proporcionado sea mayor a 0. Si el usuario ingresa un radio no válido, el programa debe mostrar un mensaje de error correspondiente. <br>
Recuerda utilizar la fórmula del área de un círculo: A=πr2, donde r representa el radio del círculo y π es el valor aproximado de 3.14159. Para usar π, puedes importar la librería math y llamarlo: math.pi<br>

```python
import math

def calculoArea(radio):
    # Calcula el área del círculo utilizando la fórmula: Área = π * radio^2
    return math.pi * (radio ** 2)

def main():
    try:
        radio = float(input("Escribe el radio del círculo que quieres saber (debe ser mayor a 0): "))
        if radio <= 0:
            print("Error: El radio debe ser un número mayor a 0.")
        else:
            area = calculoArea(radio)
            print(f"El área del círculo con radio {radio:.2f} es: {area:.2f}")
    except ValueError:
        print("Eso no es posible! El radio debe ser mayor a 0!")

if __name__ == "__main__":
    main()
```

## ListaDeCompra
***
Se desea crear un programa en Python que gestione una lista de compras de un usuario. El programa debe permitir al usuario mediante un menu: <br> <br>
1. Agregar elementos: El usuario debe poder ingresar nuevos productos a la lista de compras, especificando tanto el nombre del producto como la cantidad deseada. Si el producto ya está insertado, debe mostrar un mensaje indicando que el producto ya ha estado insertado y volver al menú principal.  <br>
2. Eliminar elementos: Se mostrarán el listado de elementos que haya actualmente en la lista. El usuario debe poder eliminar productos de la lista de compras mediante el número que tenga en la lista.  <br>
1. Mostrar lista: El usuario debe poder ver la lista de compras completa, con la cantidad de cada producto. En caso que la lista esté vacía, debe mostrar un mensaje diciéndolo.  <br>
1. Salir: El programa debe ofrecer una opción para salir del programa.  <br>

```python
#definición de funciones  metodosPpropios.py
def agregar_producto(lista_compras, producto, cantidad):

    for item in lista_compras:
        if item["producto"].lower() == producto.lower():
            print(f"El producto '{producto}' ya está en la lista de compras. Debe apetecerte mucho")
            return
    lista_compras.append({"producto": producto, "cantidad": cantidad})
    print(f"'{producto}' añadido correctamente. Necesitas {cantidad}.")

def eliminar_producto(lista_compras, indice):
    try:
        producto_eliminado = lista_compras.pop(indice - 1)
        print(f"'{producto_eliminado['producto']}' ha sido eliminado de la lista. No se lo diremos a nadie")
    except IndexError:
        print("Índice inválido. Intenta nuevamente.")

def mostrar_lista(lista_compras):
    
    if not lista_compras:
        print("La lista de compras está vacía.")
        return
    print("Lista de compras:")
    for i, item in enumerate(lista_compras, start=1):
        print(f"{i}. {item['producto']} - Cantidad: {item['cantidad']}")
```

```python
#menu listaDeCompra.py utiliza las funciones definidas en el codigo anterior.

def menu():

    lista_compras = []

    while True:
        print("\n--- Menú Lista de Compras ---")
        print("1. Escribe el nombre del producto que quieres comprar")
        print("2. Borra el producto de la lista")
        print("3. Ver lista")
        print("4. Salir")
        
        try:
            opcion = int(input("Elige una opción: "))
        except ValueError:
            print("Por favor, introduce un número válido si quieres ir a comprar o escribe 4 para salir.")
            continue

        if opcion == 1:
            producto = input("Que quieres comprar?' Escribe el producto: ").strip()
            try:
                cantidad = int(input("Cuantos necesitas?: "))
                if cantidad > 0:
                    agregar_producto(lista_compras, producto, cantidad)
                else:
                    print("La cantidad debe ser mayor que 0.")
            except ValueError:
                print("Alguien no quiere ir a comprar. Introduce una cantidad válida.")
        elif opcion == 2:
            mostrar_lista(lista_compras)
            try:
                indice = int(input("Introduce el número del producto a eliminar: "))
                eliminar_producto(lista_compras, indice)
            except ValueError:
                print("Introduce un número válido.")
        elif opcion == 3:
            mostrar_lista(lista_compras)
        elif opcion == 4:
            print("Saliendo del programa. ¡Hasta luego!")
            break
        else:
            print("Opción no válida. Por favor, elige entre 1 y 4.")

```

```python
# Ejecutar el programa llamando al menú solo despues de haber ejecutado previamente los dos anteriores
menu()
```

## MasterMindPY
***
Escribe un programa que genere aleatoriamente un valor entero entre 1 y 20 para a continuación pedir al usuario que lo acierte introduciendo el 
valor que crea que puede ser. <br>
En caso de error se indicará si el valor que se busca es mayor o más pequeño que el último introducido y se pedirá de nuevo introducir un valor posible y el número de intentos que le quedan. <br>
En total se tienen 5 intentos y en caso de agotarse se indicará que se ha perdido.  <br>
En caso de acierto, se indicará que se ha ganado y se mostrará el total de intentos  utilizados.  <br>
Para generar un valor aleatorio a Python deberás importar la librería random al inicio de tu script vía la sentencia import random. <br>
Con la librería ya importada, para generar un valor entero aleatorio dentro de un rango solo hay que invocar el método random.randint(min,max). <br>
Para facilitar el ejercicio se mostrará por pantalla el número aleatorio.<br>

```python
import random

def juego_adivinar_numero():

    numero_secreto = random.randint(1, 20)
    intentos = 5 
    intentos_usados = 0

    print("¡Bienvenido al juego Ayuda a Medusin!")
    print("Medusin ha olvidado el pin de su ordenador y debes ayudarle a entrar.\nPara ello debes encontrar el codigo de 4 dígitos que lo desbloquea. ")
    print("Son numeros entre 1 y 20. Pero cuidado solo tienes 5 intentos para adivinarlo o el ordenador se bloqueara.")

    while intentos > 0:
        try:
        
            numero_usuario = int(input("Introduce tu número: "))
            intentos_usados += 1

            if numero_usuario < 1 or numero_usuario > 20:
                print("Recuerda, ha de ser un número entre 1 y 20.")
                continue

        
            if numero_usuario == numero_secreto:
                print(f"¡Felicidades! Has acertado el número {numero_secreto} en {intentos_usados} intentos.\nEl ordenador se ha desbloqueado y Medusin te lo agradece con un cafe Premium.")
                return
            elif numero_usuario < numero_secreto:
                print(f"El número secreto es mayor que {numero_usuario}.")
            else:
                print(f"El número secreto es menor que {numero_usuario}.")
        except ValueError:
            print("Recuerda, ha de ser un número entre 1 y 20.")
            continue

        intentos -= 1
        print(f"Te quedan {intentos} intento(s).")

    print(f"Lo siento, has perdido. El número secreto era {numero_secreto}. Medusin llora desconsoladamente :( )")


juego_adivinar_numero()
```
