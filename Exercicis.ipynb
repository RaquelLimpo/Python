### M7T7 1era versió 
Crea una llista amb 10 números. Un cop la tinguis creada, recorre la llista i imprimeix el sumatori 
de tots els números i la mitjana d'aquests.<br>
Fes l'exercici sense utilitzar funcions de python com sum(), la idea és practicar els bucles.<br>

```python
print("Benvingut usuari/ària al programa que demana 10 números, els suma i en fa mitjana")

numeros = []

#  fem un bucle per cada numero generat per range, que en aquest cas comença per 1 i acaba en 11 perque range es menja l'ultim
for i in range(1, 11):
    num = int(input(f"Introdueix el número {i}: "))
    numeros.append(num)  # Afegir el número introduït a la llista buida

suma_total = 0

for num in numeros:
    suma_total += num  # Afegeix el valor del número a la suma total

mitjana = suma_total / len(numeros)

print("El sumatori de tots els números és:", suma_total)
print("La mitjana dels números és:", mitjana)

print("fi del programa")
```

### M7T9
Crea un programa que demani 5 articles i el seu preu.<br>

### M7T10
Escriu un programa que demani un text i retorni un diccionari amb les lletres que apareixen 
i la quantitat de cops que apareixen.

```python
print("Aquest programa demana un text i retorna un diccionari amb les lletres que apareixen i la quantitat de cops que apareixen.")
text = input("Introdueix un text: ")

comptador_lletres = {}

for lletra in text:
        lletra = lletra.lower()  # Conveteix a minúscules
        if lletra in comptador_lletres:
            comptador_lletres[lletra] += 1  # Augmenta el comptador
        else:
            comptador_lletres[lletra] = 1  # Afegeix al diccionari

print("\nDiccionari de lletres i quantitat de cops que apareixen:")
print(comptador_lletres)

print("fi del programa")
```
