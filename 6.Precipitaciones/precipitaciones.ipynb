## IncrementarPrecipitaciones
***

Hacer un programa en Python que le a los valores de precipitaciones que hay en un fichero de texto llamado “precipitaciones.txt”  <br>
Esto lo debes hacer mediante un menú como el que se muestra a continuación: <br>
Cada opción del menú se debe realizar usando métodos. <br>
Se deben usar los siguientes métodos: <br>
• leerDatos: Es método que va a leer los valores de las precipitaciones del fichero y los almacenará en una lista. <br>
La función va a devolver la lista generada al menú principal. Esta función se encargará de abrir y cerrar el fichero en modo lectura. <br>
• mostrarDatos: Es un método que mostrará por pantalla los valores leídos anteriormente.<br>
En caso de que no se hayan leído o la lista esté vacía, debe mostrar el mensaje: “No hay datos cargados” <br>  
Si no se han leído los datos, debe mostrar lo siguiente: <br>
En caso contrario, se muestran los datos cargados en la lista. <br>
• incrementarPrecipitaciones: Es un método que le pasaremos la lista con los valores de las precipitaciones leídas. <br>
El método preguntará al usuario el valor a incrementar. Este valor debe ser validado entre 1 y 200 <br>
 Este método va a incrementar cada valor de precipitación con el valor a incrementar. <br>
En caso de que no se hayan leído o la lista esté vacía, debe mostrar el mensaje: <br>
“No hay datos cargados” <br>
Ejemplo de un valor fuera del rango entre 1 y 200. Después se pone correctamente. <br>
• almacenarPrecipitacionesFichero: Le pasaremos la lista con los valores de las precipitaciones incrementadas y lo vamos a guardar en un fichero cuyo nombre será preguntado al usuario.  <br>
Damos por supuesto que el usuario pondrá la extensión .txt correctamente en el nombre del fichero.  <br>
Esta función se encarga de volcar los datos ya modificados en la lista. <br>
En caso de que no se hayan leído o la lista esté vacía, debe mostrar el mensaje: <br>
“No hay datos cargados” <br> 
• pintaMenu: Mostrará el menú preguntando la opción a escoger. Debe devolver la opción escogida.  <br>


```python
from metodosPropios import (
    leerDatos, mostrarDatos, incrementarPrecipitaciones, 
    almacenarPrecipitacionesFichero, pintaMenu
)
nombre_fichero = "precipitaciones.txt"
lista_precipitaciones = []

while True:
    opcion = pintaMenu()
    if opcion == 1:
        lista_precipitaciones = leerDatos(nombre_fichero)
    elif opcion == 2:
        mostrarDatos(lista_precipitaciones)
    elif opcion == 3:
        lista_precipitaciones = incrementarPrecipitaciones(lista_precipitaciones)
    elif opcion == 4:
        almacenarPrecipitacionesFichero(lista_precipitaciones)
    elif opcion == 5:
        print("Saliendo del programa.")
        break
    else:
        print("Opción no válida. Inténtalo de nuevo.")
```
